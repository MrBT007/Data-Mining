```python
# Pipeline
# drop constant and duplicate at a time
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('constant',DropConstantFeatures(tol=0.99)),
                  ('Duplicate', DropDuplicateFeatures())])
xnew = pipe.fit_transform(x)
```

```python
xnew.shape
```

```python
xnew
```

```python
pipe.named_steps['constant'].features_to_drop_
```
