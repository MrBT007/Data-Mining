```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data['Income'] =  data['Income'].fillna(data['Income'].mean())
# data.Income.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data['Education'] = enc.fit_transform(data['Education'])
data['Marital_Status'] = enc.fit_transform(data['Marital_Status'])
data['Dt_Customer'] = enc.fit_transform(data['Dt_Customer'])
```

```python
data
```

```python
data['Year_Birth'].value_counts()
```

```python
# moto data hoy to dimension +1 = n_sample thay
x = data.drop('ID',axis=1)
x
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_new = sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_new = pca.fit_transform(x_new)
x_new
```

```python
plt.scatter(x_new[:,0],x_new[:,1])
plt.show()
```

```python
x.nunique(axis=0)
```

```python
from sklearn.cluster import KMeans
n = []
for i in range(1,11):
    model = KMeans(n_clusters=i , random_state=0)
    model.fit(x_new)
    n.append(model.inertia_)
plt.plot(range(1,11),n)
plt.title("Elbow method")
plt.xlabel("Number of cluster")
plt.ylabel("WCC")
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 
dendogram = sch.dendrogram(sch.linkage(x_new,method='ward'))
plt.xlabel('No of clusters')
plt.ylabel('Euclidean Dist')
```
