```python
new_data = data.loc[data['Country'] == 'France']
new_data
```

```python
t = [str(new_data['Description'].iloc[i]).split() for i in range(8557)]
t
```

```python
from apyori import apriori
```

```python
rules = apriori(t,min_support=0.002,min_confidence=0.5,min_lift=3,max_length=2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultDataFrame
```

```python
new_df = resultDataFrame.nlargest(n=10,columns='Lift')
new_df
```
