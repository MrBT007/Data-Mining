```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("House_price2.csv")
data
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
data = data.dropna(axis=1)
```

```python
x = data.drop(['SalePrice'],axis=1)
y = data['SalePrice']
```

```python
from feature_engine.selection import DropDuplicateFeatures , DropConstantFeatures , DropCorrelatedFeatures
from sklearn.pipeline import Pipeline

pip = Pipeline([('Constant',DropConstantFeatures(tol=0.98)),('Duplicate',DropDuplicateFeatures()),('Corr',DropCorrelatedFeatures())])
```

```python
x_new = pip.fit_transform(x);
x_new
```

```python
categorical_cols = [cname for cname in x_new.columns if x_new[cname].nunique() < 10 and x_new[cname].dtype == 'object']
numerical_cols = [cname for cname in x_new.columns if x_new[cname].dtype in ['int64','float64']]
x_new = x_new[categorical_cols + numerical_cols]
x_new
```

```python
x_new.columns
```

```python
for i in categorical_cols:
  x_new[i] = enc.fit_transform(x_new[i])
```

```python
pip = Pipeline([('Constant',DropConstantFeatures(tol=0.98)),('Duplicate',DropDuplicateFeatures()),('Corr',DropCorrelatedFeatures())])
x_new = pip.fit_transform(x_new);
x_new
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred1 = model.predict(x_test)
mean_absolute_error(y_test,y_pred1)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.09,n_jobs = 4)
my_model.fit(x_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(x_test, y_test)], 
             verbose=False)

y_predxg = my_model.predict(x_test)
mean_absolute_error(y_test,y_predxg)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)

x_new=poly.fit_transform(x_new)

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()

regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)

mean_absolute_error(y_test,y_pred)
```
