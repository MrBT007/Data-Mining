```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("marketing_campaign.csv",sep='\t')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)  
data['Income'] =  data['Income'].fillna(data['Income'].mean())
# data.Income.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data['Education'] = enc.fit_transform(data['Education'])
data['Marital_Status'] = enc.fit_transform(data['Marital_Status'])
data['Dt_Customer'] = enc.fit_transform(data['Dt_Customer'])
```

```python
data
```

```python
data.nunique()
```

```python
data.drop(['Year_Birth', 'ID', 'Z_CostContact', 'Z_Revenue', 'Dt_Customer', 'Marital_Status','Response','Complain','AcceptedCmp2','AcceptedCmp1','AcceptedCmp5','AcceptedCmp4','AcceptedCmp3'],axis=1,inplace=True)
```

```python
x_new = data
x_new
```

```python
from sklearn.cluster import KMeans
n = []
for i in range(1,11):
    model = KMeans(n_clusters=i , random_state=0)
    model.fit(x_new)
    n.append(model.inertia_)
plt.plot(range(1,11),n)
plt.title("Elbow method")
plt.xlabel("Number of cluster")
plt.ylabel("WCC")
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
silhouette_scores = []
for i in range(2,10):
    m1=AgglomerativeClustering(n_clusters=i)
    c = m1.fit_predict(x_new)
    silhouette_scores.append(silhouette_score(x_new, m1.fit_predict(x_new))) 
plt.bar(range(2,10), silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('S(i)', fontsize = 20) 
plt.show()
```

```python
sc=max(silhouette_scores)
number_of_clusters=silhouette_scores.index(sc)+2
print("Number of Cluster Required is : ", number_of_clusters)
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 
dendogram = sch.dendrogram(sch.linkage(x_new,method='ward'))
plt.xlabel('No of clusters')
plt.ylabel('Euclidean Dist')
```
